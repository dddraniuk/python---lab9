```python
import numpy as np
import matplotlib.pyplot as plt


plt.style.use('seaborn-v0_8-whitegrid')
```

### Система рівнянь (варіант №6)
\begin{cases}
-10.3x_1 + 10.2x_2 = 70 \\
4.7x_1 + 12.3x_2 = 173 \\
13.2x_1 + 8.8x_2 = 282
\end{cases}

```python
a1, b1, c1 = -10.3, 10.2, 70
a2, b2, c2 = 4.7, 12.3, 173
a3, b3, c3 = 13.2, 8.8, 282

x1 = np.linspace(0, 30, 200)
x2_1 = (c1 - a1 * x1) / b1
x2_2 = (c2 - a2 * x1) / b2
x2_3 = (c3 - a3 * x1) / b3
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
fig.suptitle("Окремі графіки трьох прямих", fontsize=14)

axes[0].plot(x1, x2_1, color='red', linestyle='--')
axes[0].set_title(r'$-10.3x_1 + 10.2x_2 = 70$')
axes[0].set_xlabel('x₁')
axes[0].set_ylabel('x₂')

axes[1].plot(x1, x2_2, color='purple', linestyle='-.')
axes[1].set_title(r'$4.7x_1 + 12.3x_2 = 173$')
axes[1].set_xlabel('x₁')

axes[2].plot(x1, x2_3, color='green', linestyle=':')
axes[2].set_title(r'$13.2x_1 + 8.8x_2 = 282$')
axes[2].set_xlabel('x₁')

plt.tight_layout()
plt.show()
```

```python
def intersection(a1, b1, c1, a2, b2, c2):
    A = np.array([[a1, b1],
                  [a2, b2]])
    B = np.array([c1, c2])
    return np.linalg.solve(A, B)

p12 = intersection(a1, b1, c1, a2, b2, c2)
p13 = intersection(a1, b1, c1, a3, b3, c3)
p23 = intersection(a2, b2, c2, a3, b3, c3)

p12, p13, p23
```

```python
plt.figure(figsize=(8, 16), dpi=100)
plt.title("Усі прямі на одному рисунку (варіант 6)", fontsize=14)

plt.plot(x1, x2_1, 'r--', label=r'$-10.3x_1 + 10.2x_2 = 70$')
plt.plot(x1, x2_2, 'm-.', label=r'$4.7x_1 + 12.3x_2 = 173$')
plt.plot(x1, x2_3, 'g:', label=r'$13.2x_1 + 8.8x_2 = 282$')

plt.scatter(p12[0], p12[1], color='blue', s=50)
plt.scatter(p13[0], p13[1], color='blue', s=50)
plt.scatter(p23[0], p23[1], color='blue', s=50)

plt.text(p12[0] - 3, p12[1] + 3, 'Точка перетину (L1, L2)', fontsize=9)
plt.text(p13[0] - 3, p13[1] + 3, 'Точка перетину (L1, L3)', fontsize=9)
plt.text(p23[0] - 3, p23[1] + 3, 'Точка перетину (L2, L3)', fontsize=9)

X = [p12[0], p13[0], p23[0]]
Y = [p12[1], p13[1], p23[1]]
plt.fill(X, Y, color='gray', alpha=0.3, label='Область перетину')

plt.text(5, (c1 - a1*5)/b1 - 5, r'$-10.3x_1 + 10.2x_2 = 70$', fontsize=8, color='red')
plt.text(10, (c2 - a2*10)/b2 + 2, r'$4.7x_1 + 12.3x_2 = 173$', fontsize=8, color='magenta')
plt.text(15, (c3 - a3*15)/b3 - 5, r'$13.2x_1 + 8.8x_2 = 282$', fontsize=8, color='green')

plt.xlabel("x₁")
plt.ylabel("x₂")
plt.grid(True, linestyle=':', color='gray')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 16), dpi=100)
plt.title("Збереження рисунку у форматах PNG, JPG, SVG", fontsize=14)

plt.plot(x1, x2_1, 'r--', label='L1')
plt.plot(x1, x2_2, 'm-.', label='L2')
plt.plot(x1, x2_3, 'g:', label='L3')

X = [p12[0], p13[0], p23[0]]
Y = [p12[1], p13[1], p23[1]]
plt.fill(X, Y, color='gray', alpha=0.3)

plt.xlabel("x₁")
plt.ylabel("x₂")
plt.legend()
plt.grid(True, linestyle=':')

plt.savefig("LABORATORNA9.png", dpi=100)
plt.savefig("LABORATORNA9.jpg", dpi=100)
plt.savefig("LABORATORNA9.svg", dpi=100)
plt.show()
```
